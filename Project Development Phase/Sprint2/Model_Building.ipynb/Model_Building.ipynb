```python
#Converting the categorical to 0 and 1
```

```python
df['Red_Blood_Cells'].value_counts()
```

```python
df['Red_Blood_Cells'] = df['Red_Blood_Cells'].replace(to_replace = {'normal' : 0, 'abnormal' : 1})
```

```python
df['Pus_Cells'].value_counts()
```

```python
df['Pus_Cells'] = df['Pus_Cells'].replace(to_replace = {'normal' : 0, 'abnormal' : 1})
```

```python
df['Puss_Cell_Clumps'].value_counts()
```

```python
df['Puss_Cell_Clumps'] = df['Puss_Cell_Clumps'].replace(to_replace = {'notpresent':0,'present':1})
```

```python
df['Hypertension'].value_counts()
```

```python
df['Hypertension'] = df['Hypertension'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

```python
df['Diabetes_Mellitus'].value_counts()
```

```python
df['Diabetes_Mellitus'] = df['Diabetes_Mellitus'].replace(to_replace = {'\tyes':'yes', ' yes':'yes', '\tno':'no'})
```

```python
df['Coronary_Artery_Disease'].value_counts()
```

```python
df['Coronary_Artery_Disease'] = df['Coronary_Artery_Disease'].replace(to_replace = {'\tno':'no'})
```

```python
df['Appetite'].unique()
```

```python
df['Appetite'] = df['Appetite'].replace(to_replace={'good':1,'poor':0,'no':np.nan})
```

```python
df['Peda_Edema'].value_counts()
```

```python
df['Peda_Edema'] = df['Peda_Edema'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

```python
df['Aanemia'].value_counts()
```

```python
df['Aanemia'] = df['Aanemia'].replace(to_replace = {'yes' : 1, 'no' : 0})
```

```python
#TARGET COLUMN
df['clas_s'].value_counts()
```

```python
df['clas_s'] = df['clas_s'].replace(to_replace={'ckd\t':'ckd'})
```

```python
df["clas_s"] = [1 if i == "ckd" else 0 for i in df["clas_s"]]
```

```python
# Gonna remove the bacteria column , nothing is present in the dataset 
df.drop ('Bacteria', axis= 1, inplace = True)
```

```python
df.info() #Converting PCV,WBCC,RBCC ( Objects to numberic)
```

```python
df['Packed_Cell_Volume'] = pd.to_numeric(df['Packed_Cell_Volume'], errors='coerce')
df['White_Blood_Cell_Count'] = pd.to_numeric(df['White_Blood_Cell_Count'], errors='coerce')
df['Red_Blood_Cell_Count'] = pd.to_numeric(df['Red_Blood_Cell_Count'], errors='coerce')
```

```python
df.info()
```

⚫Handling The Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
cat_features = ['Red_Blood_Cells','Pus_Cells','Puss_Cell_Clumps',
              'Hypertension','Diabetes_Mellitus','Coronary_Artery_Disease',
              'Appetite','Peda_Edema','Aanemia']
```

```python
for feature in cat_features:
    df[feature] = df[feature].fillna(df[feature].mode())
```

```python
num_features = ['Age','Blood_Pressure','Specific_Gravity','Albumin','Sugar',
              'Blood_Gulcose_Random','Blood_Urea','Serum_Creatinine','Sodium','Potassium','Haemoglobin',
              'White_Blood_Cell_Count','Red_Blood_Cell_Count']
```

```python
for feature in num_features:
    df[feature] = df[feature].fillna(df[feature].mode())
```

⚫LABEL ENCODING

```python

from sklearn.preprocessing import LabelEncoder

lab_enc = LabelEncoder()
for i in df:
  df[i] = lab_enc.fit_transform(df[i])
```

⚫Splitting The Dataset Into Dependent And Independent Variable

```python
# Independent and Dependent Feature:
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

⚫CHECKING FEATURE IMPORTANCE

```python
# Feature Importance:
from sklearn.tree import DecisionTreeClassifier
model =  DecisionTreeClassifier()
model.fit(X,y)

plt.figure(figsize=(8,6))
ranked_features=pd.Series(model.feature_importances_,index=X.columns)
ranked_features.nlargest(24).plot(kind='barh')
plt.show()
```

```python
ranked_features.nlargest(8).index
```

```python
X = df[['Haemoglobin', 'Specific_Gravity', 'Red_Blood_Cell_Count', 'Albumin', 'Blood_Urea', 'Blood_Pressure', 'Blood_Gulcose_Random', 'Serum_Creatinine']]
X.head()
```

⚫SPLITTING THE DATASET INTO TRAINSET AND SPLIT TEST

```python
# Train Test Split: 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=33)
print(X_train.shape)
print(X_test.shape)
```

⚫BUILDING THE MODEL

```python
# Importing Performance Metrics:
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# DecisionTreeClassifier:
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model = model.fit(X_train,y_train)
```

⚫TESTING THE MODEL

```python
# Predictions:
y_pred = model.predict(X_test)
y_pred = model.predict([[10, 20 , 30 , 40 , 50 , 60 , 80 , 80]])
print(y_pred)
```

⚫MODEL EVALUATION

```python
y_pred = model.predict(X_test)
# Performance:
print('Accuracy:', accuracy_score(y_test,y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_pred))
print('Classification Report:')
print(classification_report(y_test,y_pred))
```

⚫SAVING THE MODEL

```python
#Saving the model 
pickle.dump(model,open('_CKD_NEW2.pkl','wb'))
df.to_csv('processed_kidney_disease2.csv', index=False)
```
